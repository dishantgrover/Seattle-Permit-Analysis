## Creating a neural network by using categorical as well as some continuous variables 

```python
dep_var = 'Category'
cat_names = ['Permit','Action','Status','Work']
cont_names = ['Latt','Long']
procs = [FillMissing, Categorify, Normalize]
```

```python
test = TabularList.from_df(df.copy(), cat_names=cat_names,cont_names = cont_names)
```

```python
data = (TabularList.from_df(df,  cat_names=cat_names,cont_names=cont_names,procs=procs)
                           .split_by_rand_pct(valid_pct=0.2)
                           .label_from_df(cols=dep_var)
                           .add_test(test)
                           .databunch())
```

```python
type(data)
```

```python
data.show_batch(rows=10)
```

```python
learn = tabular_learner(data, layers=[200,100], metrics=accuracy)
```

```python
learn.fit(5, 1e-5)
```
